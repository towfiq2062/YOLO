```python
#WORKING FINE_SAVES CONFUSION MATrix and reports


import os
import csv
import pandas as pd
import numpy as np
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sn
import matplotlib.pyplot as plt

# Define paths
base_run_dir = r'/mnt/nrdstor/amilab/towfiqami/Final/codes/runs/classify'
truth_dir = r'/mnt/nrdstor/amilab/towfiqami/Final/csvs/truth'
preds_dir = r'/mnt/nrdstor/amilab/towfiqami/Final/csvs/preds'
merged_dir = r'/mnt/nrdstor/amilab/towfiqami/Final/csvs/merged'
confusion_dir = r'/mnt/nrdstor/amilab/towfiqami/Final/csvs/Confusion matrix'

# Prediction folder mappings
predictions = {
    'predict': ('data', 'data'),
    'predict2': ('T1', 'T1'),
    'predict3': ('T1', 'T2'),
    'predict4': ('T1', 'T3'),
    'predict5': ('T2', 'T1'),
    'predict6': ('T2', 'T2'),
    'predict7': ('T2', 'T3'),
    'predict8': ('T3', 'T1'),
    'predict9': ('T3', 'T2'),
    'predict10': ('T3', 'T3'),
    'predict11': ('HL1', 'HL1'),
    'predict12': ('HL1', 'HL2'),
    'predict13': ('HL2', 'HL1'),
    'predict14': ('HL2', 'HL2')

    
}

# Extract prediction label
def extract_prediction(line):
    parts = line.strip().split(' ', 1)
    return parts[1] if len(parts) == 2 else None

# Process predictions
def process_predictions(pred_key, model, test):
    input_directory = os.path.join(base_run_dir, pred_key, 'labels')
    output_path = os.path.join(preds_dir, f"{pred_key}.csv")

    if not os.path.exists(input_directory):
        print(f"Prediction directory not found: {input_directory}")
        return None

    data = []
    for filename in os.listdir(input_directory):
        if filename.endswith('.txt'):
            file_path = os.path.join(input_directory, filename)
            with open(file_path, 'r') as file:
                first_line = file.readline()
                prediction = extract_prediction(first_line)
                if prediction:
                    data.append({'filename': os.path.splitext(filename)[0], 'predict': prediction})

    os.makedirs(preds_dir, exist_ok=True)
    with open(output_path, 'w', newline='') as csv_file:
        fieldnames = ['filename', 'predict']
        writer = csv.DictWriter(csv_file, fieldnames=fieldnames)
        writer.writeheader()
        writer.writerows(data)

    print(f"Predictions saved for {pred_key}: {output_path}")
    return output_path

# Merge predictions and truth
def merge_predictions_with_truth(pred_csv, model, test):
    truth_file = os.path.join(truth_dir, f"{test}_truth.csv")
    if not os.path.exists(truth_file):
        print(f"Truth file not found: {truth_file}")
        return None

    predict_df = pd.read_csv(pred_csv)
    truth_df = pd.read_csv(truth_file)
    merged_df = pd.merge(truth_df[['filename', 'truth']], predict_df[['filename', 'predict']], on='filename')

    merged_output_path = os.path.join(merged_dir, f"{model}_model_tested_on_{test}_data.csv")
    os.makedirs(merged_dir, exist_ok=True)
    merged_df.to_csv(merged_output_path, index=False)
    print(f"Merged CSV saved: {merged_output_path}")
    return merged_output_path

# Generate confusion matrix and classification report
def generate_confusion_and_report(merged_csv, model, test, classification_report_list):
    df = pd.read_csv(merged_csv)
    truth, predict = df['truth'], df['predict']
    classes = sorted(truth.unique())

    # # Confusion matrix (counts)
    # conf_matrix = confusion_matrix(truth, predict, labels=classes)
    # conf_matrix_normalized = conf_matrix.astype('float') / conf_matrix.sum(axis=1)[:, np.newaxis]

    # # Plot confusion matrix
    # plt.figure(figsize=(12, 10))
    # annotated_matrix = [[f"{int(count)}\n({normalized:.2f})" for count, normalized in zip(row, row_normalized)] 
    #                     for row, row_normalized in zip(conf_matrix, conf_matrix_normalized)]
    # sn.heatmap(conf_matrix_normalized, annot=annotated_matrix, fmt='', cmap='Blues',
    #            xticklabels=classes, yticklabels=classes, cbar_kws={'label': 'Normalized Value'})
    # plt.xlabel('Predicted')
    # plt.ylabel('Actual')
    # plt.title(f'Confusion Matrix: {model} model on {test} data')

    # # Save confusion matrix as image
    # os.makedirs(confusion_dir, exist_ok=True)
    # confusion_matrix_path = os.path.join(confusion_dir, f"{model}_model_on_{test}_confusion_matrix.png")
    # plt.savefig(confusion_matrix_path)
    # plt.close()
    # print(f"Confusion matrix saved: {confusion_matrix_path}")
    

# Compute confusion matrix (counts)
    conf_matrix = confusion_matrix(truth, predict)
    
    # Normalize confusion matrix
    conf_matrix_norm = conf_matrix.astype('float') / conf_matrix.sum(axis=1)[:, np.newaxis]
    
    # Plot confusion matrix
    plt.figure(figsize=(8, 6))
    sn.set(font_scale=1)  # for label size
    
    # Create heatmap
    ax = sn.heatmap(conf_matrix_norm, annot=False, fmt='.2f', cmap='Blues', 
                    xticklabels=classes, yticklabels=classes)
    
    # Annotate heatmap with both normalized values and raw counts
    for i in range(conf_matrix.shape[0]):
        for j in range(conf_matrix.shape[1]):
            normalized_value = conf_matrix_norm[i, j]
            raw_value = conf_matrix[i, j]
            text = "{:.2f}\n{}".format(normalized_value, raw_value)
            color = 'black' if normalized_value < 0.5 else 'white'  # Adjust text color for better readability
            ax.text(j + 0.5, i + 0.5, text, ha='center', va='center', color=color)
    
    #plt.title(f'Confusion Matrix: {model} model on {test} data')
    plt.xlabel("Predicted Label")
    plt.ylabel("True Label")
    # Save the confusion matrix plot
    os.makedirs(confusion_dir, exist_ok=True)
    confusion_matrix_path = os.path.join(confusion_dir, f"{model}_model_on_{test}_confusion_matrix.png")
    plt.savefig(confusion_matrix_path, dpi=300, bbox_inches='tight')
    plt.close()
    
    # Print the file path of the saved confusion matrix
    print(f"Confusion matrix saved: {confusion_matrix_path}")

    # Classification report
    class_report = classification_report(truth, predict, output_dict=True, labels=classes)
    
    # Add class-specific metrics
    for label, metrics in class_report.items():
        if label in classes:  # Include only class-specific metrics
            classification_report_list.append({
                'Model': model,
                'Test Data': test,
                'Class': label,
                'Precision': metrics['precision'],
                'Recall': metrics['recall'],
                'F1-Score': metrics['f1-score'],
                'Support': metrics['support']
            })
    
    # Add overall metrics (accuracy, macro avg, weighted avg)
    classification_report_list.append({
        'Model': model,
        'Test Data': test,
        'Class': 'accuracy',
        'Precision': None,  # Accuracy does not have precision
        'Recall': None,     # Accuracy does not have recall
        'F1-Score': None,   # Accuracy does not have F1-score
        'Support': class_report['accuracy']  # Accuracy as a single value
    })
    
    if 'macro avg' in class_report:
        classification_report_list.append({
            'Model': model,
            'Test Data': test,
            'Class': 'macro avg',
            'Precision': class_report['macro avg']['precision'],
            'Recall': class_report['macro avg']['recall'],
            'F1-Score': class_report['macro avg']['f1-score'],
            'Support': class_report['macro avg']['support']
        })
    
    if 'weighted avg' in class_report:
        classification_report_list.append({
            'Model': model,
            'Test Data': test,
            'Class': 'weighted avg',
            'Precision': class_report['weighted avg']['precision'],
            'Recall': class_report['weighted avg']['recall'],
            'F1-Score': class_report['weighted avg']['f1-score'],
            'Support': class_report['weighted avg']['support']
        })

# Main workflow
classification_report_list = []

for pred_key, (model, test) in predictions.items():
    print(f"\nProcessing: {model} model tested on {test} data")

    pred_csv = process_predictions(pred_key, model, test)
    if not pred_csv:
        continue

    merged_csv = merge_predictions_with_truth(pred_csv, model, test)
    if not merged_csv:
        continue

    generate_confusion_and_report(merged_csv, model, test, classification_report_list)

# Save all classification reports
classification_report_path = os.path.join(confusion_dir, "classification_reportsFor_OWN_test.csv")
classification_report_df = pd.DataFrame(classification_report_list)
classification_report_df.to_csv(classification_report_path, index=False)
print(f"\nAll classification reports saved: {classification_report_path}")

print("\nProcessing complete.")
```
